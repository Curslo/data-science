```python
# Barplot of a categorical column
sns.countplot(data=df, x='Donation')

# Histogram of a numeric column
df['Health_Score'].hist(bins=20)
plt.title("Outcome")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```
